```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as sc
from scipy.interpolate import lagrange as l
import numpy.polynomial.polynomial as npp
```

```python
x = toch["arr_0"]
y = toch["arr_1"]
```

```python
f = l(x,y)
```

```python
npp.Polynomial(f).coef
```
